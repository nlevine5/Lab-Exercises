```python
words = ["apple", "chestnut", "gargoyle", "pandas", "sheep", "raptor"]

def longest(listofwords):
    length = 0
    
    for i in listofwords:
        newlength = len(i)
        if newlength > length:
            length = newlength 
            longest = i
            
    
    return longest

longest(words)
        

    
```

```python
#dictionaries begin with assigning a dictionary name and using these: {}

example = {}

#you can add to or change a dictionary this way

example = {"key1":"val1"}
print(example)

example["key2"] = "val2"
print(example)

example["key1"] = "changed"
print(example)

#here are some useful functions

#.update() can also be used to add to a dictionary

example.update({"key3":"updated"})
print(example)

#.get() can be used to get the value from a specific key

getkey = example.get("key1")
print(getkey)

#.pop() can remove the specified key

example.pop("key1")
print(example)

#.clear() will clear the dictionary

example.clear()
print(example)
```

```python
#a system to manage a students school record

class Student:
    def __init__(self, name, age, grade, mathgrades, enggrades, histgrades, scigrades, violation):
        self.name = name
        self.age = age
        self.grade = grade
        self.mathgrades = mathgrades
        self.enggrades = enggrades
        self.histgrades = histgrades
        self.scigrades = scigrades
        self.violation = violation
        
    def student(self):
        print(f"{self.name}, {self.age} years old, grade {self.grade}")
        
    def getgrades(self, subject, avg_or_all):
        if subject == "math":
            if avg_or_all == "avg":
                print(sum(self.mathgrades)/len(mathgrades))
            elif avg_or_all == "all":
                print(self.mathgrades)
        if subject == "english":
            if avg_or_all == "avg":
                print(sum(self.enggrades)/len(self.enggrades))
            elif avg_or_all == "all":
                print(self.enggrades)
        if subject == "history":
            if avg_or_all == "avg":
                print(sum(self.histgrades)/len(self.histgrades))
            elif avg_or_all == "all":
                print(self.histgrades)
        if subject == "science":
            if avg_or_all == "avg":
                print(sum(self.scigrades)/len(self.scigrades))
            elif avg_or_all == "all":
                print(self.scigrades)
            
    def changegrade(self, subject, grades, add_or_remove):
        if subject == "math":
            if add_or_remove == "add":
                self.mathgrades.append(grades)
                print(self.mathgrades)
            elif add_or_remove == "remove":
                self.mathgrades.remove(grades)
                print(self.mathgrades)
        if subject == "english":
            if add_or_remove == "add":
                self.enggrades.append(grades)
                print(self.enggrades)
            elif add_or_remove == "remove":
                self.enggrades.remove(grades)
                print(self.enggrades)
        if subject == "history":
            if add_or_remove == "add":
                self.histgrades.append(grades)
                print(self.histgrades)
            elif add_or_remove == "remove":
                self.histgrades.remove(grades)
                print(self.histgrades)
        if subject == "science":
            if add_or_remove == "add":
                self.scigrades.append(grades)
                print(self.scigrades)
            elif add_or_remove == "remove":
                self.scigrades.remove(grades)
                print(self.scigrades)
                
    def studentrecord(self, violation, see_or_add):
        if see_or_add == "see":
            print(self.violation)
        elif see_or_add == "add":
            if len(self.violation) < 3:
                self.violation.append(violation)
                print(self.violation)
            elif len(self.violation) >= 3:
                print(f"{self.name} already has 3 violations. Due to their recent violation of {violation}, they are suspended.")
            
s1 = Student("Nicolas Levine", 16, 11, [77, 33, 92], [91, 99, 83], [78, 81, 82], [90, 74, 83], [])
s1.student()
s1.getgrades("science", "all")
s1.getgrades("science", "avg")
s1.changegrade("science", 77, "add")
s1.changegrade("science", 74, "remove")
s1.getgrades("science","avg")
s1.studentrecord("Fighting", "add")
s1.studentrecord("Late", "add")
s1.studentrecord("Skipped Class", "add")
s1.studentrecord("Vaping in class", "add")
```
